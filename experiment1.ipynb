```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Create a Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# How many Iris versicolor flowers are in the dataset?
versicolor_count = df[df['species'] == 'versicolor'].shape[0]
print("Number of Iris versicolor flowers:", versicolor_count)

# What is the average petal length of all Iris flowers?
average_petal_length = df['petal length (cm)'].mean()
print("Average petal length of all Iris flowers:", average_petal_length)

# What is the average sepal length of all Iris flowers?
average_sepal_length = df['sepal length (cm)'].mean()
print("Average sepal length of all Iris flowers:", average_sepal_length)

# What is the correlation between petal length and petal width?
correlation = df['petal length (cm)'].corr(df['petal width (cm)'])
print("Correlation between petal length and petal width:", correlation)
```
